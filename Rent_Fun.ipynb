```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
```

```python
### Annual household income (lowest)
incomeLow = 150 # in units of $1,000's

### Annual household income (highest)
incomeHigh = 300 # in units of $1,000's

### Monthly rent (lowest)
rentLow = 3000 # per month

### Monthly rent (highest)
rentHigh = 5000 # per month
```

```python
steps = 100

income_range = np.multiply([incomeLow,incomeHigh], 1000)

income = np.ones((steps,1))
annualRent = np.ones((1,steps))

income_space = np.linspace(income_range[0], income_range[1], steps)
income[:,0] = np.divide(1, income_space)

annualRent[0,:] = np.linspace(rentLow*12, rentHigh*12, steps)

ratioData = np.matmul(income,annualRent)

xlabels = np.array(['$'+ str(x)[:1] + ',' + str(x)[1:4] for x in np.linspace(rentLow, rentHigh, 9)])
ylabels = np.array(['$'+ str(x)[:3] +'K' for x in np.linspace(income_range[0], income_range[1], 9)//1000])

fig, ax = plt.subplots(figsize=(12,12))
im = ax.imshow(np.rot90(ratioData.T), cmap = 'turbo')#, vmin = 0, vmax = 1)
ax.set_xticks(np.linspace(0, steps, 9))
ax.set_yticks(np.linspace(0, steps, 9))
ax.set_xticklabels(xlabels)
ax.set_yticklabels(np.flip(ylabels))
plt.colorbar(im)
plt.grid()
plt.title('Rent as Percentage of Annual Income')
plt.xlabel('Monthly Rent')
plt.ylabel('Annual Household Income')

plt.show()
```
